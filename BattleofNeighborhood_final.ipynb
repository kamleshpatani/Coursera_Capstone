```python
#function to repeat the same process to all area
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Area', 
                  'Area Latitude', 
                  'Area Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)

#run the above function on each neighborhood and create a new dataframe
bmumbai_venues = getNearbyVenues(names=bmumbai['Area'],
                                   latitudes=bmumbai['Latitude'],
                                   longitudes=bmumbai['Longitude']
                                  )

#check the size of the resulting dataframe
print(bmumbai_venues.shape)
bmumbai_venues.head()
```

<h1>Analyze Mumbai</h1>

```python
# one hot encoding
bmumbai_onehot = pd.get_dummies(bmumbai_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
bmumbai_onehot['Area'] = bmumbai_venues['Area'] 

# move neighborhood column to the first column
fixed_columns = [bmumbai_onehot.columns[-1]] + list(bmumbai_onehot.columns[:-1])
bmumbai_onehot = bmumbai_onehot[fixed_columns]

#examine the new dataframe size after one hot encoding
print('{} rows were returned after one hot encoding.'.format(bmumbai_onehot.shape[0]))

#group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
bmumbai_grouped = bmumbai_onehot.groupby('Area').mean().reset_index()

#examine the new dataframe size after one hot encoding
print('{} rows were returned after grouping.'.format(bmumbai_grouped.shape[0]))
```

```python
#print each neighborhood along with the top 5 most common venues
num_top_venues = 5

for hood in bmumbai_grouped['Area']:
    print("----"+hood+"----")
    temp = bmumbai_grouped[bmumbai_grouped['Area'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#put into a pandas dataframe

#write a function to sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

#create the new dataframe and display the top 10 venues for each neighborhood
num_top_venues = 8

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Area']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
areas_venues_sorted = pd.DataFrame(columns=columns)
areas_venues_sorted['Area'] = bmumbai_grouped['Area']

for ind in np.arange(bmumbai_grouped.shape[0]):
    areas_venues_sorted.iloc[ind, 1:] = return_most_common_venues(bmumbai_grouped.iloc[ind, :], num_top_venues)

areas_venues_sorted.head()
```

```python
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 3

bmumbai_grouped_clustering = bmumbai_grouped.drop('Area', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(bmumbai_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 

#create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
bmumbai_merged = bmumbai.drop(index=6)

# add clustering labels
bmumbai_merged['Cluster Labels'] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
bmumbai_merged = bmumbai_merged.join(areas_venues_sorted.set_index('Area'), on='Area')

bmumbai_merged.head()
```
